**1) Proportion sous nutri**

Filtrons entre l'année 2016 et 2018

```python
An = '2016-2018'
filtre_data = data_nutri[(data_nutri['Année'] == An)]
filtre_data
```

```python
filtre_data["Valeur"].astype("float")
```

```python
dt = pd.DataFrame(filtre_data)

# Convertir la colonne en valeurs numériques
dt['Valeur'] = pd.to_numeric(dt['Valeur'], errors='coerce')

# Applique la fonction de formatage avec une expression lambda
#dt['Valeur'] = dt['Valeur'].apply(lambda x: f'< {x:.2f}' if x < 0.1 else f'{x:.2f}' if not pd.isna(x) else '')

# Affiche le DataFrame résultant
print(dt)
```

```python
dt["Valeur"] = dt["Valeur"].fillna(0)
```

Calculons la somme totale pour cette année

```python
total_population = dt["Valeur"].sum()
total_population
```

```python
proportion = 100*(1e6 * total_population) / (7*1e9)

print("La proportion de la population mondiale en 2017 est de: {:.2f}%".format(proportion))
```

**2) Nombre théorique**

Filtrons sur la dispo kcal/per/j

```python
dispo = data_dispo['Disponibilité alimentaire (Kcal/personne/jour)']
```

```python
data_dispo
```

```python
prod = data_dispo*365*dispo['Valeur'] #produit 
somme = prod.sum() #somme
nb = (somme / 2500*365).round() #nbre de personne nourrit
print('\n le nombre des personnes qui pourraient être nourries est de :', nb)
```
